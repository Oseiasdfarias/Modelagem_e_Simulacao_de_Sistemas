# **<center> Erro em Regime Permanente Sistema de Primeira Ordem<center/>**

Considerando um sistema $G(s)$

$$G(s) = \dfrac{K}{\tau S + 1 }$$

<br>

>> Onde

>> $K$: Ganho do Sistema \
>> $\tau$ Constante de Tempo
Controlador Proporcional $C(s)$

<br>

$$C(s) = K_p$$

<br>

Obs: Vamos considerar um Sensor unitário para esse sistema, ou seja, o sinal de saida será realimentado no comparador sem alteração de escala.



$R(s)$ é o sinal aplicado ao sistema, para nossa análise, iremos usar o Degrau Unitário $U(s) = \dfrac{1}{S}$

<br>

## **<center>Equação de erro para um sistema de primeira ordem com realimentação<center/>**

<br>

O erro para um sistema realimentado é dado pelo sinal de entrada menos o sinal de realimentação, como mostra a equaçõa abaixo:

<br>

$$E(s) = R(s) - Y(s)$$

<br>

O sinal de saída $Y(s)$ é dado pela multiplicação do sinal $U(s)$ pelo sistema $G(s)$, logo:

<br>

$$E(s) = R(s) - U(s)\cdot G(s)$$

<br>

Ainda, podemos observar que o sinal $U(s)$ é dado pela multiplicação do erro $E(s)$ pelo controlador proporcional $C(s)$, e assim, temos:

<br>

$$E(s) = R(s) - G(s) \cdot C(s) \cdot E(s)$$

<br>

Agora podemos encontrar a função de transferência para o erro em estado permanente.

<br>

$$E(s) = R(s) - G(s) \cdot C(s) \cdot E(s)$$ \

$$E(s) + G(s) \cdot C(s) \cdot E(s)= R(s)$$ \

$$[1 + G(s) \cdot C(s)] E(s)= R(s)$$ \

$$\dfrac{E(s)}{R(s)} = \dfrac{1}{1 + G(s) \cdot C(s)}$$ \

<br>

**<center> EQUAÇÃO DO ERRO<center/>** 

<br>

$$E(s) = \dfrac{1}{1 + G(s) \cdot C(s)}R(s)$$ \

<br>


Para encontrar o valor do Erro no estado estacionário $e(\infty)$, vamos usar o Teorema do Valor Final, onde;

## **<center>Teorema do valor final<center/>**

<br>

$$f(\infty) = \lim_{S\to 0} sF(s)$$

<br>

## **<center>Encontrando o ganho $K$ em regime permanente $y(\infty)$<center/>**

Vamos primeiramente aplicar esse teorema na função de transferência em malha aberta, para encontrar o valor de saída em Regime Permanente $Y(\infty)$.

$$G(s) = \dfrac{K}{\tau S + 1 }$$ \

$$U(s) = \dfrac{1}{S}$$ \

$$\dfrac{Y(s)}{U(s)} = \dfrac{K}{\tau S + 1 }$$ \

$$Y(s) = \dfrac{K}{\tau S + 1 } U(s)$$ \

$$Y(s) = \dfrac{K}{\tau S + 1 } \dfrac{1}{S}$$

<br>

Aplicando o teorema do valor final, na função de transferência em malha aberta, temos:

$$y(\infty) = \lim_{S\to 0} S \dfrac{K}{\tau S + 1 } \dfrac{1}{S}$$

$$y(\infty) = \lim_{S\to 0} \dfrac{K}{\tau S + 1 }$$

Substituindo $0$ em $S$

$$y(\infty) = \dfrac{K}{0 + 1 }$$ \

$$y(\infty) =  K$$ \

Assim, podemos concluir que o $\bf{K}$ nos dá o valor em regime permanente do nosso sistema. ou seja, o valor de regime permanente do sistema.

<br>

## **<center>Encontrando o erro em regime permanente $e(\infty)$<center/>**

Aplicando o Teorema de Valor Final na equação do erro com o controlador proporcional, para encontrar o valor do erro em Regime Permanente $e(\infty)$

<br>

**Equação Geral**

$$e(\infty) = \lim_{S\to 0} S\dfrac{1}{1 + G(s) \cdot C(s)}R(s)$$ \



Vamos aplicar essa teoria em um exemplo prático.

<br>

**Exemplo para um Controlador Proporcional**

$$G(s) = \dfrac{K}{\tau S + 1 }$$ \

$$U(s) = \dfrac{1}{S}$$ \

$$C(s) = K_p$$ \

$$e(\infty) = \lim_{S\to 0} S \cdot \dfrac{1}{1 + K_p \cdot \dfrac{K}{\tau S + 1 }} \cdot \dfrac{1}{S}$$ \

Aplicando as manipulaçoes algêbricas, temos

$$e(\infty) = \lim_{S\to 0} \dfrac{1}{1 + K_p \cdot \dfrac{K}{\tau S + 1 }}$$ \

$$e(\infty) = \dfrac{1}{1 + K_p K}$$ \


<br>

A partir dessa solução, podemos comcluir que o nosso controlador proporcional não é capaz de zerar o erro $e(\infty)$ em regime permanente do nosso sistema.

O que pode ser feito para atenuar esse erro é almentar o valor de $K_p$, no entando sabemos que os sistemas reais saturam, além de que, valores muito altos de $K_p$ torna o sistema instável.

# **<center>Aplicação Prática<center/>**

### **<center>Erro em Regime Permanente<center/>**

<br>


$$K_p = 2$$ 
$$\tau = 1,2 s$$ 
$$K = 0,8$$ \

$$e(\infty) = \dfrac{1}{1 + 2 \cdot 0,8} = 0,3846$$ \

$$K_p = 20$$ 
$$\tau = 1,2 s$$ 
$$K = 0,8$$ \

$$e(\infty) = \dfrac{1}{1 + 20 \cdot 0,8} = 0,0588$$ \

```python
import control as ctl 
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

## **Usando o K = 2**

```python
# Função de Transferência do erro
K_p = 2 
tau = 1.2 
K = 0.8   # Ganho do sistema
Temp_simu = 3.0

# Ganho do Sistema em Regime Permanente com Controlador Proporcional
y_inf = (K_p*K)/(1+(K_p*K))

# Erro estacionário do sistema, com Controlador Proporcional
e_inf = 1/(1+K_p*K)
print(f"Erro em Regime Permanente: {e_inf}")


# Função de Transferência do Controlador Proporcional
C_s = ctl.TransferFunction(K_p, 1)
#print(C_s)

# Função de Transferência da Planta
G_s = ctl.TransferFunction(K, [tau, 1])
#print(G_s)

# Cria a Função de Transferência do Sensor
P_s = ctl.TransferFunction([1.],[1.])

# Função de Transferência do Controlador Proporcional em série com a Planta
H_s = ctl.series(C_s, G_s)
print(H_s)

F_s = ctl.feedback(H_s, P_s, sign=-1);
print(f"\n> Função de Transferência em Malha fechada:\n {F_s}")

# Resposta da Função de Transferência do Erro ao Degrau
t1, y1 = ctl.step_response(F_s, Temp_simu)

# Função de Transferência do Erro
E_s = 1 / (1 + H_s)
print("\n\nFunção de Transferência do Erro E(s)\n", E_s)

# Resposta da Função de Transferência do Erro ao Degrau
t, y = ctl.step_response(E_s, Temp_simu)

# Plotando o degrau unitário
Temp_deg = np.linspace(-0.2, Temp_simu, 1000)
degrau = np.ones_like(Temp_deg)
degrau[Temp_deg < 0] = 0

plt.figure(figsize = ((15, 6)))
plt.suptitle("Gráficos", fontweight="bold", fontsize = 20)

ax1 = plt.subplot(1, 2, 1)
plt.plot(t1, y1)
plt.plot(Temp_deg, degrau, color = "gray", linewidth = 2.)

ax1.plot([ y_inf*4, 0], [y_inf, y_inf],color='black', lw=1.2)
ax1.text(0.01, y_inf+0.05, f"y$(\infty)$ = {y_inf}",fontsize=14)

# Eixos x, y
ax1.axhline(0, color='black', lw=1.2)
ax1.axvline(0, color='black', lw=1.2)

plt.title("Resposta ao Degrau para Sistema", fontweight="bold", fontsize = 14)
plt.xlabel("Tempo $(t)$", fontweight="bold", fontsize = 14)
plt.ylabel("Resposta y(t)", fontweight="bold", fontsize = 14)

# Plotagem do gráfico do erro
ax = plt.subplot(1, 2, 2)
plt.plot(t, y)
plt.plot(Temp_deg, degrau, color = "gray", linewidth = 2.)

ax.plot([ tau*2.5, 0], [e_inf, e_inf],color='black', lw=1.2)
ax.text(0.01, e_inf-0.05, f"e$(\infty)$ = {e_inf}",fontsize=14)

# Eixos x, y
ax.axhline(0, color='black', lw=1.2)
ax.axvline(0, color='black', lw=1.2)

plt.title("Resposta ao Degrau - Função de Erro - Sistema de 1° Ordem", fontweight="bold", fontsize = 14)
plt.xlabel("Tempo $(t)$", fontweight="bold", fontsize = 14)
plt.ylabel("Erro e(t)", fontweight="bold", fontsize = 14)

plt.show()
```
