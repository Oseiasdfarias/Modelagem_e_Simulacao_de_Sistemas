#**Modelagem Matemático e simulação do Sistema (Massa Mola Amortecedor)**

1.png

Para esse sistema, podemos utilizar a **segunda Leis de Newton** para modela-lo.

Sabenos que a força resultante $f_R$ é igual ao somatório das forças que atual em com corpo.

<br>

$$f_R = \sum\limits_{}^{}m\ddot{x}$$

<br>

Se aplicarmos uma força externa $f_e(t)$ em nosso sistema e considerarmos ela positiva, podemos definir uma força na mola $f_m(t)$ e no amortecedor $f_a(t)$ que se opoem a essa força externa.

<br>

> $f_e(t)$: Força externa aplicada

> $f_m(t)$: Força que atua na Mola (Lei de Hooke - Mola ideal) 

> $f_a(t)$: Força que atua no Amortecedor (Amortecedor ideal)

<br>

Sabenos ainda que, temos a posição $x(t)$ e sua derivada nos dá a velocidade $\dot{x}(t)$ e sua segunda derivada nos dá a aceleração do nosso sistema $\ddot{x}(t)$.

<br>

**Resumo:**

> Posição (Saída do sistema): $x(t)$

> Velocidade: $\dot{x}(t)$

> Aceleração do sistema: $\ddot{x}(t)$


Dessa forma, obtemos a seguinte equação que modela nosso sistema:

<br>

$$m\ddot{x}(t) = f(t) - f_m(t) - f_a(t)$$

<br>

$$f_m(t) = K_mx(t)$$

$$f_a(t) = K_a\dot{x}(t)$$

<br>

$$m\ddot{x}(t) = f(t) - K_mx(t) - K_a\dot{x}(t)$$

<br>

Dessa forma, podemos encontrar uma **EDO** que discreve esse sistema.

<br>

$$m\ddot{x}(t) + K_mx(t) + K_a\dot{x}(t) = f(t)$$

<br>

Temdo obtido a **EDO** que discreva esse sistema, podemos usar a Transformada de Laplace para trabalhar com esse sistema no domínio da frequência.

<br>

$$mS^2\dot{x}(s) + K_aSx(s) + K_mx(s) = f(s)$$

<br>

Agora que temos nosso sistema no domínio da frequência, podemos obter a função de transferência.

<br>

$$(mS^2 + K_aS + K_m)x(s) = F(s)$$

$$G_s = \dfrac{x(s)}{F(s)} = \dfrac{1}{mS^2 + K_aS + K_m}$$

<br>

Função de Transferência:

<br>

$$G_s = \dfrac{1}{mS^2 + K_aS + K_m}$$

<br>

A partir da **Função de Transferência** podemos simular a resposta do sistema ao degrau $u(t)$ para observamos o comportamento do sistema, para isso vamos usar a Biblioteca Control do Python.

#**Simulação do Sistema Usando Python**

```python
import matplotlib.pyplot as plt
import control as ct 

plt.style.use('ggplot')
```

```python
m = 1      #  Massa do nosso sistema
k_m = 4    #  Constante elástica da mola
k_a = 0.9  #  Constante do amartecedor

num = [1]            # Numerador da Função de Transferêrncia
den = [1, k_a, k_m]  # Denominador da Função de Transferêrncia

# Crinando a Função de Transferência com auxílio da biblioteca Control
G_s = ct.TransferFunction(num, den)

# Simulando a resposta ao degrau 
yout, T = ct.step_response(G_s)

# Plotando o gráfico da resposta ao degrau
plt.figure(figsize=(12, 6))
plt.plot(yout, T, linewidth= 2)
plt.title("Sistema Massa Mola Amortecedor", fontsize=17)
plt.xlabel("Tempo", fontsize=15)
plt.ylabel("Resposta", fontsize=15)
plt.show()
```

```python
m = 1      #  Massa do nosso sistema
k_m = 4    #  Constante elástica da mola
k_a = 2  #  Constante do amartecedor

num = [1]            # Numerador da Função de Transferêrncia
den = [1, k_a, k_m]  # Denominador da Função de Transferêrncia

# Crinando a Função de Transferência com auxílio da biblioteca Control
G_s = ct.TransferFunction(num, den)

# Simulando a resposta ao degrau 
yout, T = ct.step_response(G_s)

# Plotando o gráfico da resposta ao degrau
plt.figure(figsize=(12, 6))
plt.plot(yout, T, linewidth= 2)
plt.title("Sistema Massa Mola Amortecedor", fontsize=17)
plt.xlabel("Tempo", fontsize=15)
plt.ylabel("Resposta", fontsize=15)
plt.show()
```
